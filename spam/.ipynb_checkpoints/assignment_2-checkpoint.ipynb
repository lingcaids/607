```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from pandas import read_csv
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = read_csv("spam.csv").values
X = data[:, :57]
y = data[:, 57]
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.2)
C = DecisionTreeClassifier()
C.fit(X_tr, y_tr)
y_p = C.predict(X_te)
print(sum(y_p == y_te)/len(y_p))
```

```python
C = BaggingClassifier(n_estimators=100, oob_score=True)
C.fit(X_tr, y_tr)
print(f"OOB score: {C.oob_score_}")
y_p = C.predict(X_te)
print(sum(y_p == y_te)/len(y_p))
```

```python
C = RandomForestClassifier(n_estimators=100, oob_score=True)
C.fit(X_tr, y_tr)
print(f"OOB score: {C.oob_score_}")
y_p = C.predict(X_te)
print(sum(y_p == y_te)/len(y_p))
plt.barh(range(len(C.feature_importances_)), np.sort(C.feature_importances_))
plt.show()
```

```python
scores = []
for k in range(1, 20):
    C = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(C, X, y, cv=5, scoring="accuracy").mean()
    scores.append(score)

print(np.argmax(scores) + 1)
plt.plot(range(1, 20), scores)
plt.show()
# X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.2)
# C = KNeighborsClassifier(n_neighbors=2)
# C.fit(X_tr, y_tr)
# y_p = C.predict(X_te)
# print(sum(y_p == y_te)/len(y_p))
```
