```python
from __future__ import absolute_import, division, print_function
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
model = tf.keras.Sequential()
model.add(layers.Dense(2, input_dim=2, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',
              optimizer='rmsprop', # work here
              metrics=['accuracy'])
```

```python
w1 = np.array([[20, -20], [20, -20]])
b1 = np.array([-30, 10])
w2 = np.array([[20], [20]])
b2 = np.array([-10])
model.set_weights((w1, b1, w2, b2))
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
model.predict(X)
```

```python
model.get_weights()
```

```python
n = 2**10
X = np.random.randint(2, size=(n, 2)) + np.random.normal(0, 0.2, size=(n, 2))
y = model.predict_classes(X).reshape(-1)
plt.plot(X[y == 0, 0], X[y == 0, 1], 'ro' )
plt.plot(X[y == 1, 0], X[y == 1, 1], 'bo' )
plt.show()
```
