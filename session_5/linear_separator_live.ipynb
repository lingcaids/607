```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
```

```python
np.random.seed(666)
theta = np.random.uniform(0, 2*np.pi)
normal_vector = np.array([np.cos(theta), np.sin(theta)])
x0 = np.random.uniform(-1, 1, size=40)
x1 = np.random.uniform(-1, 1, size=40)
X = np.array([x0, x1]).T
I = np.abs(X.dot(normal_vector)) > 0.1
X = X[I]
y = np.sign(X.dot(normal_vector))

plt.figure(figsize=(6, 6))
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.plot(X[y < 0, 0], X[y < 0, 1], 'ro', markeredgecolor='k', markersize=9)
plt.plot(X[y > 0, 0], X[y > 0, 1], 'bo', markeredgecolor='k', markersize=9)

t = np.arange(-1.2, 1.2, 0.1)
# plt.plot(t, -normal_vector[0]/normal_vector[1]*t, 'k')
plt.show()
```

```python
def loss(theta, x, y):
    return max(-y*x.dot(np.array([np.cos(theta), np.sin(theta)])), 0)

def slope(theta, x, y):
    if loss(theta, x, y) == 0:
        return 0
    else:
        return -y*(x[0]*(-np.sin(theta)) + x[1]*np.cos(theta))
```

```python
h = 0.05 # learning rate
theta_ns = []
theta = 1
for i in range(13):
    m = np.sum([slope(theta, xi, yi) for (xi, yi) in zip(X, y)])
    theta_ns.append(theta)
    theta = theta - h*m
```
