## Problem 1

Your colleague Sally has developed a new regression technique. Unfortunately, she's not familiar with `sklearn`'s API, so she's implemented her technique in two functions, `fit` and `predict`, not as a python class. These functions are defined in file `fit_and_predict.py`. Import them.

- `fit` takes two positional arguments, one-dimensional vectors `X` and `y` of the same length (training data) and one keyword argument `k`, a positive integer model hyperparameter. The function `fit` returns an object meant to be passed to `predict` as its the keyword argument `fit_data`.

- `predict` takes one positional argument, a one-dimensional vector `X` (test data), and one keyword argument `fit_data`. The keyword argument `fit_data` should be passed the output from the `fit` function. The function `predict` outputs a vector `y` of predicted $y$-values corresponding to the input vector `X`.

Your main task is to wrap her functions in a python class that implements the usual `sklearn` regressor API.

You'll then be using this class to analyze the data contained in `data.csv`. Its first column is the feature (there's only one) and the second column is the target. To load it, use `np.loadtxt("data.csv", delimiter=",")`.

### Your tasks:
1. Split the provided dataset into training and testing sets. Then use cross validation on the training set to determine the optimal value of `k`. With that parameter value, fit Sally's model to the training data. 

2. Using your class, predict the $y$-values for `x = np.arange(-4.25, 4.25, 0.05)`. Then plot these predicted values versus `x` to see the regression curve. 

```python
from fit_and_predict import fit, predict
from sklearn.base import BaseEstimator, RegressorMixin
from sklearn.model_selection import cross_val_score
from matplotlib import pyplot as plt
import numpy as np

class Klassifier(BaseEstimator, RegressorMixin):
    
    def __init__(self, k=None):
        if k is None:
            raise Exception("Required keyword argument k is missing.")
        self.k = k
        
    def fit(self, X, y):
        self.X_ = X
        self.y_ = y
        self.fit_data = fit(X, y, k=self.k)
        return self
        
    def predict(self, X):
        return predict(X, fit_data=self.fit_data)
    
B = np.loadtxt("data.csv", delimiter=",")
X = B[:, 0]
y = B[:, 1]
    
scores = []
for k in range(1, 10):
    C = Klassifier(k=k)
    score = cross_val_score(C, X, y, cv=5, scoring="neg_mean_squared_error")
    scores.append(score.mean())

C = Klassifier(k=np.argmax(scores) + 1)
C.fit(X, y)

t = np.arange(-4.25, 5.25, 0.1)
y_pred = C.predict(t)

plt.plot(t, y_pred)
plt.show()
```

## Problem 2
In this problem, we'll be working with the famous <a href="https://scikit-learn.org/stable/datasets/index.html#iris-dataset">iris flower dataset</a>. The goal is identify an iris as one of three species based on physical measurements. This dataset is included with `sklearn`; import `load_iris` from the `sklearn.datasets` module.

1. Split the iris dataset into training and testing sets, `X_train`, `y_train` and `X_test`, `y_test`, containing 2/3 and 1/3 of the data points, respectively.
2. Train a 6-nearest neighbors classifier and use it to predict the class labels of the data points in the testing set.
3. For each pair of features `i` and `j`, make a scatter plot of `X_test[i]`, `X_test[j]`. The color of the marker should indicate the predicted class label of the point. Misclassified points should be black and be plotted with `markersize=10`, so they stand out.

```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import numpy as np

data = load_iris()
X = data.data
y = data.target
X_train, X_test, y_train, y_test = train_test_split(X, y)
C = KNeighborsClassifier(n_neighbors=6)
C.fit(X_train, y_train)
y_pred = C.predict(X_test)
print(f"Accuracy: {sum(y_pred == y_test)/len(y_pred):.2f}")

markers = ['ro', 'bo', 'go']
f, ax = plt.subplots(2, 3, sharey=True, figsize=(12, 8))
count = 0
for i in range(3):
    for j in range(i+1, 4):
        a = ax[count % 2][count % 3]
        for k in range(3):
            a.plot(X_test[y_pred == k, i], X_test[y_pred == k, j], markers[k])
        a.plot(X_test[y_pred != y_test, i], X_test[y_pred != y_test, j], 'ko', markersize=10)
        a.set_xlabel(data.feature_names[i])
        a.set_ylabel(data.feature_names[j])  
        count = count + 1
```
