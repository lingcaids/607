```python
from sklearn.neighbors import KNeighborsRegressor, RadiusNeighborsRegressor
from sklearn.model_selection import cross_val_score
import numpy as np
from matplotlib import pyplot as plt
```

```python
def r(x):
    """A convenience function to compute values of r."""
    return 0.5*x*np.cos(4*np.pi*x)

def make_y(x):
    """A convenience function to sample from N(r(x), 0.1)."""
    return r(x) + 0.1*np.random.normal(size=len(x))
```

```python
from sklearn.model_selection import LeaveOneOut

# np.random.seed(666)
n = 50
X = np.sort(np.random.uniform(size=n))
y = make_y(X)

loo = LeaveOneOut()

all_ys = []
scores = []
for k in range(1, 20):
    R = KNeighborsRegressor(n_neighbors=k)
    losses = []
    ys = []
    for train_ind, test_ind in loo.split(X):
        X_train = X[train_ind]
        X_test = X[test_ind]
        y_train = y[train_ind]
        y_test = y[test_ind]
        R.fit(X_train.reshape(-1, 1), y_train)
        yi_ = R.predict(X_test.reshape(-1, 1))
        ys.append(yi_[0])
        losses.append((yi_ - y_test)**2)
    scores.append(np.mean(losses))
    all_ys.append(ys)
    
```

```python
plt.plot(range(1, 20), scores)
print(np.argmin(scores))
```
