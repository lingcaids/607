```python
from sklearn.datasets import load_breast_cancer, load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = load_breast_cancer()
X = data.data
y = data.target
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.25)
C = DecisionTreeClassifier()
C.fit(X_tr, y_tr)
plt.figure(figsize=(18, 12))
ax = plt.gca()
ax.set_title("Relative feature importance", fontsize=24)
ax.barh(data.feature_names, C.feature_importances_)
ax.set_yticklabels(data.feature_names, fontsize=18)
plt.show()
```

```python
importances = []
C = DecisionTreeClassifier()
for j in range(200):
    X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.25)
    C.fit(X_tr, y_tr)
    importances.append(C.feature_importances_)
imps = np.mean(importances, axis=0)
plt.figure(figsize=(18, 12))
ax = plt.gca()
ax.set_title("Average relative feature importance", fontsize=24)
ax.barh(data.feature_names, imps)
ax.set_yticklabels(data.feature_names, fontsize=18)
plt.show()
```
